## import fire data

```r
clFiles <- list.files(path=firefolder, pattern='*.-CL.tif')
jdFiles <- list.files(path=firefolder, pattern='*.-JD.tif')
clst <- raster::stack(as.list(paste0(firefolder,clFiles)))
jdst <- raster::stack(as.list(paste0(firefolder,jdFiles)))
#head(clFiles)
#head(jdFiles)
```

## Create shapefile with fire locations

```r
shapefile(SpatialPoints(crds, crs(clst)),paste0(ofolder, tsfile, 'fireLoc'),overwrite=T)
```

## Extract fire dates

```r
extrjd <- extract(jdst,crds)
extrcl <- extract(clst,crds)
```

```r
#extrjd
cli <- as.matrix(extrcl)
jdi <- as.matrix(extrjd)
nmi <- lapply(names(jdst),substr, 2, 9)
dts <- as.Date(unlist(nmi),'%Y%m%d')
yrs <- unlist(lapply(names(jdst),substr, 2, 5))  

fdate <- list()
for(i in 1:dim(cli)[1]){
    fdate[[i]] <- c(crds[i,],as.Date(paste0(as.numeric(yrs[cli[i,]>95 & jdi[i,]>0]),'-',jdi[i,cli[i,]>95 & jdi[i,]>0]),'%Y-%j'))
}    
```

## Export

```r
save(fdate, file = file.path(ofolder,paste0(tsfile, 'fireDate.rda')))
```

## plot time series together with fire date

```r
head(fdate)

for( i in 1:100){
    tsii <- as.numeric(tsi[i,-c(1,2)])
    tmii <- as.Date(names(tsi[-c(1,2)]))
    plot(zoo(tsii[is.na(tsii)==F], tmii[is.na(tsii)==F]),
        xlab='Time [year]', ylab = 'NBR [-]')
    #lines(zoo(cli[i,is.na(cli[i,])==F]/100, dts[is.na(cli[i,])==F]), col='green')
    abline(v=(fdate[[i]][-c(1,2)]), col = 'red', lw=3, lty=5)
}
```
